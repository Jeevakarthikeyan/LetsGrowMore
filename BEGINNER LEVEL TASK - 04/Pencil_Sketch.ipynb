#**Image to Pencil Sketch with Python**

**Importing All Required Libraries**

```python
import cv2
import matplotlib.pyplot as plt
```

**Loading Data**

**Read Photo**

```python
img=cv2.imread("GOLDENRETRIEVER.jpg")
```

```python
img
```

**Show Image using OpenCV**

```python
cv2.imshow('original image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

**Display Using Matplotlib**

```python
plt.imshow(img)
plt.axis(False)
plt.show()
```

**Matplotlib vs OpenCV**

**Convert BGR to RGB**

```python
plt.imshow(img[:,:,::-1])
plt.axis(False)
plt.show()
```

**Convert BGR to RGB**

```python
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.axis(False)
plt.show()
```

```python
image_color = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
final_image = cv2.hconcat((img,image_color))
plt.imshow(final_image)
plt.axis(False)
plt.show()
```

**Convert Image to a Pencil Sketch**

**Convert to Grey Image**

```python
grey_img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
grey_img
```

```python
cv2.imshow('Grey image',grey_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

**Invert Image**

```python
invert_img=cv2.bitwise_not(grey_img)
```

**Blur image**

```python
blur_img=cv2.GaussianBlur(invert_img, (111,111),0)
```

**Invert Blurred Image**

```python
invblur_img=cv2.bitwise_not(blur_img)
```

**Sketch**

```python
sketch_img=cv2.divide(grey_img,invblur_img, scale=256.0)
plt.imshow(sketch_img)
plt.axis(False)
plt.show()
```

**Save Sketch**

```python
cv2.imwrite('sketch.png', sketch_img)
```

**Display sketch**

```python
cv2.imshow('sketch image',sketch_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

**Original Image vs Sketch**

```python
plt.figure(figsize=(14,8))
plt.subplot(1,2,1)
plt.title('Original image', size=18)
plt.imshow(RGB_img)
plt.axis('off')
plt.subplot(1,2,2)
plt.title('Sketch', size=18)
rgb_sketch=cv2.cvtColor(sketch_img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_sketch)
plt.axis('off')
plt.show()
```
