**LabelEncoder**

```python
from sklearn.preprocessing import LabelEncoder
```

```python
Label_encoder = LabelEncoder()
```

```python
iris.Species = Label_encoder.fit_transform(iris.Species)
```

```python
iris.Species.unique()
```

```python
iris.head()
```

**Separate out your Independent feature & Dependent feature**

```python
x = iris.drop("Species",axis=1)
y = iris["Species"]
```

**Splitting Dataset into Training and Testing set**

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.25,random_state=3)
```

**K NeighborsClassifier**

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knc =KNeighborsClassifier(n_neighbors=2)
knc.fit(xtrain, ytrain)
print('Accuracy:', knc.score(xtest, ytest)*100)
```

**Prediction**

```python
y_pred_train_knc = knc.predict(xtrain) 
y_pred_train_knc
```

```python
y_pred_knc = knc.predict(xtest)
```

```python
y_pred_knc
```

**DataFrame : Actual & Predicted**

```python
df1 = pd.DataFrame({"Actual":ytest,"Predicted":y_pred_knc})
df1
```

**Confusion Matrix**

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(ytest, y_pred_knc)
```
